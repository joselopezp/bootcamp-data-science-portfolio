## 1. CRISP-DM Phase 1 — Business Understanding

### 1.1 Context

University health departments face a recurring challenge: they have limited budgets for student wellness interventions but lack data-driven criteria to decide *where to invest*. Anecdotal evidence suggests that sleep deprivation, sedentary behavior, and poor nutrition affect academic outcomes — but without statistical evidence, resource allocation remains subjective.

This study is commissioned by a **University Health & Wellbeing Department** seeking to:
- Understand which lifestyle habits most strongly predict academic performance.
- Quantify the magnitude and direction of those relationships.
- Generate evidence-based recommendations for intervention design.

> **ICI Perspective (Industrial Engineering lens):**  
> Think of the university as a production system. The student is the process. Inputs are habits (sleep, exercise, diet, study time). The output is academic performance (exam score). Identifying which inputs have the highest leverage on the output is a classic **process optimization** problem — the same logic applied in mining operations to identify bottleneck variables.

---

### 1.2 Business Question

> **Primary:** Which combination of student lifestyle habits (sleep, physical activity, diet, study time) most strongly predicts academic performance, and what is the estimated ROI of intervening on the highest-impact habit?

> **Secondary:** Are there statistically significant differences in academic performance across student subgroups (gender, part-time employment, parental education)?

## 2. Scientific Method — Step 1: Problem Definition

### 2.1 Phenomenon Under Study

University students exhibit heterogeneous academic performance (exam scores ranging from 18 to 100). A significant portion of this variance is hypothesized to be explained by **modifiable lifestyle behaviors**: sleep duration, physical activity frequency, dietary quality, and daily study hours.

Unlike socioeconomic or genetic factors, lifestyle behaviors are **actionable** — they can be targeted through university-led interventions at relatively low cost.

### 2.2 Problem Statement

**The university cannot identify which student wellness interventions to prioritize** because there is insufficient statistical evidence linking specific lifestyle habits to academic outcomes in the student population. This leads to either:
- Under-investment in high-impact habits (Type II error in decision-making), or
- Over-investment in low-impact programs (resource waste — LEAN: *muda*).

### 2.3 Scope and Boundaries

| In Scope | Out of Scope |
|----------|-------------|
| Lifestyle habits (sleep, exercise, diet, study time) | Genetic or neurological factors |
| Academic performance (exam score) | Long-term career outcomes |
| Cross-sectional snapshot analysis | Longitudinal tracking |
| Descriptive + inferential statistics | Causal inference / RCT |
| Students aged 15–30 | Graduate or doctoral students |

---
## 3. Scientific Method — Step 2: Research Questions

Research questions follow the **PICO framework** (Population, Intervention/Indicator, Comparison, Outcome), adapted for observational research:

| # | Research Question | Type |
|---|-------------------|------|
| RQ1 | Do students sleep fewer hours than the WHO-recommended minimum (7h)? | Descriptive / Inferential |
| RQ2 | Is there a statistically significant difference in exam scores between students who exercise ≥3 days/week and those who exercise less? | Comparative |
| RQ3 | Does sleep duration correlate positively with exam score? | Relational |
| RQ4 | What proportion of students display a sedentary lifestyle (<3 activity days/week), and does this differ from 50%? | Inferential (proportion) |
| RQ5 | Do students with higher diet quality score significantly better academically? | Comparative |

> **LEAN Filter:** Each RQ directly supports the primary business question. RQs without a clear link to the intervention decision were discarded to avoid analytical waste.

## 4. Scientific Method — Step 3: Hypotheses

Each hypothesis pair corresponds to a research question and a planned statistical test in Lesson 6.

---

### H1 — Sleep Duration vs. WHO Benchmark

| | |
|--|--|
| **H₀** | The mean sleep duration of students = 7 hours (μ_sleep = 7) |
| **H₁** | The mean sleep duration of students < 7 hours (μ_sleep < 7) |
| **Test** | One-sample t-test (one-tailed, left) |
| **Significance level** | α = 0.05 |
| **Business implication if H₁ accepted** | Sleep deprivation is prevalent → prioritize sleep hygiene programs |

---

### H2 — Exercise Frequency and Exam Score

| | |
|--|--|
| **H₀** | Mean exam score is equal for active (≥3 days) and sedentary (<3 days) students (μ_active = μ_sedentary) |
| **H₁** | Mean exam score is higher for active students (μ_active > μ_sedentary) |
| **Test** | Independent samples t-test (one-tailed, right) |
| **Significance level** | α = 0.05 |
| **Business implication if H₁ accepted** | Physical activity programs have measurable academic ROI |

---

### H3 — Sedentary Lifestyle Prevalence

| | |
|--|--|
| **H₀** | Proportion of sedentary students = 50% (p_sedentary = 0.50) |
| **H₁** | Proportion of sedentary students > 50% (p_sedentary > 0.50) |
| **Test** | One-sample proportion z-test (one-tailed, right) |
| **Significance level** | α = 0.05 |
| **Business implication if H₁ accepted** | Sedentarism is the majority behavior → systemic intervention needed |

---

### H4 — Diet Quality and Academic Performance

| | |
|--|--|
| **H₀** | Mean exam score is equal across diet quality groups (μ_poor = μ_fair = μ_good) |
| **H₁** | At least one diet quality group has a significantly different mean exam score |
| **Test** | One-way ANOVA (or Kruskal-Wallis if normality is violated) |
| **Significance level** | α = 0.05 |
| **Business implication if H₁ accepted** | Nutrition programs (meal plans, subsidies) may improve academic outcomes |

---

> **Note on Error Types in this context:**  
> - **Type I error (α):** Recommending an intervention that has no real effect → wasted university budget (*LEAN: muda*)  
> - **Type II error (β):** Failing to recommend an intervention that does have an effect → missed improvement opportunity (*LEAN: missed value*)  
> Setting α = 0.05 is a deliberate balance between these risks in a resource-constrained university setting.

## 6. Scientific Method — Step 5: Study Design

### 6.1 Study Type

**Observational cross-sectional study** with inferential statistical analysis.

- **Cross-sectional**: All measurements taken at a single point in time (snapshot).
- **Observational**: No manipulation of variables — students self-report their habits.
- **Inferential**: Sample statistics will be used to draw conclusions about the broader student population.

### 6.2 Sampling Strategy

The dataset represents a **convenience sample** from an online survey platform. For analytical purposes, we will treat it as a **stratified random sample** across gender and academic year, which enables subgroup comparisons.

| Sampling dimension | Value |
|--------------------|-------|
| Target population | University students aged 15–30 |
| Sample size | ~1,000 records |
| Sampling unit | Individual student |
| Stratification variables | Gender, academic year |
| Minimum required (for CLT) | n ≥ 30 per subgroup |

### 6.3 Analytical Roadmap — 6 Lessons

| Lesson | Topic | CRISP-DM Phase | Key Deliverable |
|--------|-------|----------------|----------------|
| **L1** | Scientific method & study design | Business Understanding | This notebook |
| **L2** | Probability & sampling design | Data Understanding | Probability trees, sample space |
| **L3** | Probability distributions | Data Preparation / Modeling | Distribution fitting + plots |
| **L4** | Sampling distribution & CLT | Modeling | CLT empirical verification |
| **L5** | Confidence intervals | Evaluation | CI for μ at 90%, 95%, 99% |
| **L6** | Hypothesis testing | Evaluation | Test results + business conclusions |

### 6.4 Methodological Assumptions

1. Observations are **independent** (one student does not influence another's habits).
2. The sample is **representative** of the broader student population (college-level, ages 15–30).
3. Self-reported data is assumed **accurate** — a known limitation (see Section 7).
4. The relationship between habits and performance is **approximately linear** at the population level.
5. Normality of continuous variables will be **verified empirically** in Lesson 3.

````python
# ===== Hypotheses Summary Table =====
# Structured view of all hypotheses for quick reference in the final report

def build_hypothesis_summary() -> pd.DataFrame:
    """Creates a structured summary of all study hypotheses.

    Returns:
        pd.DataFrame: Hypothesis table with H0, H1, test, alpha, and business implication.
    """
    data = {
        'ID': ['H1', 'H2', 'H3', 'H4'],
        'H0 (Null)': [
            'μ_sleep = 7h',
            'μ_active = μ_sedentary (exam score)',
            'p_sedentary = 0.50',
            'μ_poor = μ_fair = μ_good (exam score)'
        ],
        'H1 (Alternative)': [
            'μ_sleep < 7h',
            'μ_active > μ_sedentary',
            'p_sedentary > 0.50',
            'At least one group differs'
        ],
        'Statistical Test': [
            'One-sample t-test (left-tailed)',
            'Independent t-test (right-tailed)',
            'Proportion z-test (right-tailed)',
            'One-way ANOVA / Kruskal-Wallis'
        ],
        'Alpha': [0.05, 0.05, 0.05, 0.05],
        'If H1 Accepted → Action': [
            'Launch sleep hygiene program',
            'Invest in exercise facilities / incentives',
            'Systemic anti-sedentarism campaign',
            'Subsidize healthy meal plans'
        ]
    }
    return pd.DataFrame(data)


hypothesis_df = build_hypothesis_summary()
print('=== Hypothesis Summary Table ===')
print(hypothesis_df.to_string(index=False))
````
