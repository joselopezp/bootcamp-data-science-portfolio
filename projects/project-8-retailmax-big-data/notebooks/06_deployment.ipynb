## 8. Project Retrospective — CRISP-DM Cycle Complete <a id='8-project-retrospective'></a>

| Aspect | Reflection |
|---|---|
| **What worked well** | CRISP-DM structure kept the project organized across 6 notebooks. Olist dataset provided real-world complexity (nulls, type mismatches, multi-table joins). Synthetic clickstream generation with `spark.range()` was fast and scalable. |

| **Key business learnings** | 77% of customers are one-time buyers — the biggest business opportunity. High-ticket customers with low reviews (19%) represent the highest ROI for service recovery. Segmentation enables targeted marketing instead of one-size-fits-all campaigns. |
| **Business value delivered** | Scalable pipeline processing 2M+ records. Customer classification model (AUC=0.95). Four actionable customer segments with marketing recommendations. Business impact estimation quantifying ~$5.8M in opportunities. |
| **LEAN waste eliminated** | No redundant EDA (focused on decision-relevant variables). No unnecessary feature engineering (5 features sufficient for strong models). Cached only reused DataFrames (2 of 10). |

### Personal Perspective — ICI Background

This project reinforced how industrial engineering thinking maps directly to
Big Data:

- **Process optimization** → Spark's Catalyst optimizer eliminates redundant
  operations automatically, just as Lean eliminates waste in manufacturing.
- **Capacity planning** → Setting `spark.driver.memory` and partition count
  is the data equivalent of dimensioning production equipment.
- **Quality control** → The `try_cast` workaround for dirty data is analogous
  to statistical process control — catching defects before they propagate.
- **Cost-benefit analysis** → Choosing K=4 over K=3 is a business tradeoff
  (richer segmentation vs. cluster purity), not a purely statistical decision.

These are the perspectives that differentiate a data scientist with an
operations background from one with a purely technical one.
