## 1. CRISP-DM Phase 1 — Business Understanding

**Context:** Retail Insights S.A. is a consultancy serving multiple physical and
e-commerce retail chains. Their client currently segments customers using only age
and total purchase history — a low-resolution approach that prevents differentiated
marketing strategies and leaves behavioral patterns undetected.

**ICI Perspective:** From an Industrial Engineering standpoint, this is a process
improvement problem. The current segmentation process has low discriminating power —
analogous to a process with high σ variation where all units are treated identically
regardless of their actual properties. Unsupervised ML is the tool to reduce this
variation and identify stable behavioral strata, enabling targeted interventions
per stratum — exactly as SPC identifies process zones requiring different responses.

## 3. Theoretical Foundations — Unsupervised Learning

> Consigna Lección 1: Identify the main tasks solved by unsupervised learning,
> classify techniques by type, and analyze real application cases.

### 3.1 Supervised vs Unsupervised Learning

| Dimension | Supervised Learning | Unsupervised Learning |
|-----------|--------------------|-----------------------|
| **Labels** | Requires labeled target variable (y) | No labels — finds structure in X alone |
| **Goal** | Learn mapping X → y | Discover hidden patterns in X |
| **Evaluation** | Accuracy, R², RMSE vs ground truth | Silhouette, inertia, explained variance |
| **Examples** | Classification, regression | Clustering, dimensionality reduction |
| **Business use** | Predict outcome for new input | Understand structure of existing data |

**Why unsupervised here:** The client does not have pre-defined customer segments.
The goal is to discover natural groupings from the data itself — not to predict
a known label. This is the defining characteristic of unsupervised learning.

### 3.2 Main Tasks Solved by Unsupervised Learning

| Task | Definition | Business Application |
|------|-----------|---------------------|
| **Clustering** | Group similar observations together | Customer segmentation, anomaly detection |
| **Dimensionality Reduction** | Compress high-dimensional data into fewer dimensions | Visualization, noise reduction, feature extraction |
| **Association** | Find rules that describe co-occurrence of items | Market basket analysis, recommendation systems |

**This project uses:** Clustering + Dimensionality Reduction.
Association (e.g., Apriori) is out of scope — no transactional basket data available.

### 3.3 Technique Classification

#### Clustering

| Algorithm | Type | Key Characteristic | When to Use |
|-----------|------|-------------------|-------------|
| **KMeans** | Centroid-based | Minimizes within-cluster variance; requires k | Convex, roughly equal-sized clusters |
| **DBSCAN** | Density-based | Finds arbitrary-shaped clusters; detects noise | Irregular shapes, unknown k, outlier detection |
| **Hierarchical (Agglomerative)** | Connectivity-based | Builds dendrogram; no k required upfront | When visual confirmation of k is needed |

#### Dimensionality Reduction

| Algorithm | Type | Key Characteristic | When to Use |
|-----------|------|-------------------|-------------|
| **PCA** | Linear | Maximizes variance; preserves global structure | Pre-processing, feature compression, scree plots |
| **t-SNE** | Non-linear | Preserves local neighborhood structure | 2D/3D visualization of cluster separation |

### 3.4 Real Application Cases

| Industry | Technique | Application |
|----------|-----------|------------|
| Retail / E-commerce | KMeans + RFM | Customer segmentation for targeted campaigns (this project) |
| Banking | DBSCAN | Fraud detection — anomalous transactions as noise points |
| Healthcare | Hierarchical | Patient subgroup discovery for clinical trials |
| Manufacturing | PCA | Process monitoring — reduce sensor dimensionality for anomaly detection |
| Recommendation Systems | Association (Apriori) | Market basket analysis — "customers who bought X also bought Y" |

> **ICI connection:** In manufacturing, PCA on sensor arrays is equivalent to
> reducing a multivariate control chart to its principal axes — a technique
> used in process capability analysis for high-dimensional quality systems.
